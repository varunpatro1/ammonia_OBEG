```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df.head()
```

```python
filtered_df = df[df['target'].notnull()]
```

```python
filtered_df.shape
```

```python
temp = filtered_df['temp'].tolist()
target = filtered_df['target'].tolist()
```

```python
plt.scatter(temp, target, alpha=0.1)
plt.xlabel('Temperature')
plt.ylabel('NH4 Concentration')
plt.ylim(0,5)
plt.show()
```

```python
oxygen = filtered_df['o2'].tolist()
```

```python
plt.scatter(oxygen, target, alpha=0.1)
plt.xlabel('Oxygen Concentration')
plt.ylabel('NH4 Concentration')
plt.ylim(0,7)
plt.xlim(0,700)
plt.show()
```
