```python
nh4_not_null = pd.notnull(df['target'])
condensed_df = df[nh4_not_null]
print(condensed_df.shape)
```

```python
condensed_df
```

```python
ind = range(-100,400,25)
plt.hist(condensed_df['lat'], alpha = 0.5, label = 'Lattitude')
plt.hist(condensed_df['lon'], alpha = 0.5, label = 'Longitude')
plt.ylabel('Frequency')
plt.xticks(ind)
plt.xticks(rotation = 45)
plt.legend()
plt.title('Latitude and Longitude Distribution of Measurements containing NH4 Readings')
```

```python
ind = range(0,8000,500)
plt.hist(condensed_df['depth'], alpha = 0.5, label = 'Lattitude')
plt.ylabel('Frequency')
plt.xticks(ind)
plt.xticks(rotation=45)
plt.xlabel('Depth (m)')
plt.title('Depth Distribution of Measurements containing NH4 Readings')
```

```python
df.loc[df['lon'] > 180, 'lon'] = df['lon']-360
```

```python
geometry = [Point(xy) for xy in zip(df['lon'], df['lat'])]
gdf = GeoDataFrame(df, geometry = geometry)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax = world.plot(figsize = (15,15)), marker = 'o', color = 'red', markersize = 3);
```

```python
condensed_df.loc[condensed_df['lon'] > 180, 'lon'] = condensed_df['lon']-360
```

```python
geometry = [Point(xy) for xy in zip(condensed_df['lon'], condensed_df['lat'])]
gdf = GeoDataFrame(condensed_df, geometry = geometry)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax = world.plot(figsize = (15,15)), marker = 'o', color = 'red', markersize = 3);
```

```python
upper_ocean = condensed_df[condensed_df['depth'] <= 50]
print(upper_ocean.shape)
```

```python
ind = range(0,55,5)
plt.hist(upper_ocean['depth'], alpha = 0.8)
plt.ylabel('Frequency')
plt.xticks(ind)
plt.xticks(rotation=45)
plt.xlabel('Depth (m)')
plt.title('Depth Distribution in Upper Ocean')
```

```python
sub_surface = condensed_df[condensed_df['depth'] <= 300]
print(sub_surface.shape)
```

```python
ind = range(0,325,25)
plt.hist(sub_surface['depth'], alpha = 0.8)
plt.ylabel('Frequency')
plt.xlim([200,300])
plt.xticks(ind)
plt.xticks(rotation=45)
plt.xlabel('Depth (m)')
plt.title('Depth Distribution in Sub Surface Ocean')
```

```python
# trying to take out outlier nh4 concentrations
```

```python
plt.scatter(x=condensed_df['lat'],
            y=condensed_df['lon'],
            c=condensed_df['target'],
            cmap = 'plasma')
plt.colorbar(orientation='horizontal', label = 'NH4 Concentration')
```

```python
concentration_filtered = condensed_df[condensed_df['target'] < 35]
print(concentration_filtered.shape)
```

```python
plt.scatter(x=concentration_filtered['lon'],
            y=concentration_filtered['lat'],
            c=concentration_filtered['target'],
            cmap = 'coolwarm', s = 15)
plt.colorbar(orientation='horizontal', label = 'NH4 Concentration')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.rcParams["figure.figsize"] = (8,6)
```

```python
plt.hist(concentration_filtered['target'], alpha = 0.8)
plt.ylabel('Frequency')
plt.xlabel('Concentration')
```

```python
concentration_filtered_2 = condensed_df[condensed_df['target'] < 3]
print(concentration_filtered_2.shape)
```

```python
ind = [0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5, 2.75, 3]
plt.hist(concentration_filtered_2['target'],bins = 20, alpha = 0.8)
plt.ylabel('Frequency')
plt.xlabel('Concentration')
plt.xticks(ind);
plt.xticks(rotation=45)
plt.title('Concentration Distribution w/o Outliers')
```

```python
plt.scatter(x=concentration_filtered_2['lon'],
            y=concentration_filtered_2['lat'],
            c=concentration_filtered_2['target'],
            cmap = 'coolwarm', s = 15)
plt.colorbar(orientation='horizontal', label = 'NH4 Concentration')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Visualization of Concentrations in Global Ocean -- Outliers Filtered > 3')
plt.rcParams["figure.figsize"] = (8,6)
```
